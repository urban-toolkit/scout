```python
city_full = "New York, USA"
city = 'nyc'
filename = '../data/osm/pbf_files/%s.osm.pbf' % (city)
input_filename = '../data/osm/pbf_files/north-america-latest.osm.pbf'

geolocator = Nominatim(user_agent='uic')
location = geolocator.geocode(city_full).raw
 
south, north, west, east = map(float, location['boundingbox'])
bbox = f"{west},{south},{east},{north}"
```

```python
# apt get install osmium-tool

subprocess.run([
    "osmium", "extract",
    "-b", bbox,
    "-o", filename,
    "--overwrite",
    input_filename
])
```

```python
bbox = "-90.6879323,41.4150156,-90.4252158,41.6242105"
input_filename = "../data/osm/pbf_files/illinois.osm.pbf"
out_file = "../data/osm/pbf_files/waterways_bbox.osm.pbf"

bbox_output = "../data/osm/pbf_files/illinois_bbox.osm.pbf"
waterways_output = "../data/osm/pbf_files/waterways_bbox.osm.pbf"

# 1) Extract the bounding box
subprocess.run(
    [
        "osmium", "extract",
        "-b", bbox,
        "-o", bbox_output,
        "--overwrite",
        input_filename,
    ],
    check=True,
)

# 2) Filter waterways inside that bbox
subprocess.run(
    [
        "osmium", "tags-filter",
        "-o", waterways_output,
        "--overwrite",
        bbox_output,
        "w/waterway", "r/waterway",
    ],
    check=True,
)
```
